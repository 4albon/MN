```python
# Завдання 1. Вектор добутків парних чисел на попередні непарні
import numpy as np

even = np.arange(2, 24, 2)
odd = even - 1
vec = even * odd
vec
```

```python
# Завдання 2. Матриця 3×3 зі значень квадратів випадкових чисел 1–100
mat2 = np.random.randint(1, 101, (3, 3))
mat2_sq = mat2 ** 2
mat2_sq
```

```python
# Завдання 3. Матриця 5×5 з випадковими числами –2..2, середнє за модулем
mat3 = np.random.uniform(-2, 2, (5, 5))
avg_abs = np.abs(mat3).mean()
avg_abs
```

```python
# Завдання 4. Матриця 5×5 –5..5, середнє значення рядків
mat4 = np.random.uniform(-5, 5, (5, 5))
row_means = mat4.mean(axis=1)
row_means
```

```python
# Завдання 5. Два вектори по 7 елементів –10..10, вектор різниць
v1 = np.random.uniform(-10, 10, 7)
v2 = np.random.uniform(-10, 10, 7)
diff = v1 - v2
diff
```

```python
# Завдання 6. Добуток двох матриць 4×4 з цілих 7..17
M1 = np.random.randint(7, 18, (4, 4))
M2 = np.random.randint(7, 18, (4, 4))
prod_M = M1 @ M2
prod_M
```

```python
# Завдання 7. Добуток матриць 4×2 та 2×5 з цілих –5..5
A = np.random.randint(-5, 6, (4, 2))
B = np.random.randint(-5, 6, (2, 5))
res = A @ B
res
```

```python
# Завдання 8. Сума елементів побічної діагоналі матриці 2×2
X = np.random.randint(-10, 11, (2, 2))
secondary_diag_sum = X[0,1] + X[1,0]
secondary_diag_sum
```

```python
# Завдання 9. Найбільший елемент матриці 3×3 (дійсні –3..3)
mat9 = np.random.uniform(-3, 3, (3, 3))
max_elem = mat9.max()
max_elem
```

```python
# Завдання 10. Добуток матриці 3×4 і вектора 4×1
mat10 = np.random.randint(-5, 6, (3, 4))
vec10 = np.random.randint(1, 6, (4, 1))
res10 = mat10 @ vec10
res10
```

```python
# Завдання 11. Добуток матриці 2×3 і вектора 3×1
mat11 = np.random.randint(-5, 6, (2, 3))
vec11 = np.random.randint(1, 6, (3, 1))
res11 = mat11 @ vec11
res11
```
