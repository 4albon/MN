# Лабораторна робота: Операції з NumPy 1.2


## 1. Формування вектора з натуральних чисел

```python
import numpy as np
vec = np.arange(1, 6)
vec
```

## 2. Транспонування вектора

```python
row_vec = np.matrix([1,2,3,4,5])
row_vec.T
```

## 3. Розворот вектора

```python
arr = np.array([1,2,3,4,5])
arr[::-1]
```

## 4. Середнє значення

```python
np.mean(arr)
```

## 5. Мін/макс елементи

```python
arr.min(), arr.max()
```

## 6. Скалярний добуток

```python
a = np.array([1,2,3])
b = np.array([3,4,5])
np.dot(a,b)
```

## 7. Нормалізація вектора

```python
vec2 = np.array([2,4,6,8])
vec2 / np.linalg.norm(vec2)
```

## 8. Ортогональність векторів

```python
x = np.array([0,1,0])
y = np.array([1,0,0])
np.dot(x,y)
```

## 9. Унікальні значення

```python
mixed = np.array([1,1,2,3,3,4])
np.unique(mixed)
```

## 10. Арифметика з векторами

```python
p = np.array([1,3,5])
q = np.array([2,4,6])
p+q, p-q, p*q, p/q
```

## 11. Перевірка довжини вектора

```python
test_vec = np.array([1,2,3,4,5])
np.isclose(np.linalg.norm(test_vec), 5)
```

## 12. Матриця випадкових чисел

```python
m = np.random.randint(0,10,(3,3))
m
```

## 13. Визначник

```python
np.linalg.det(m)
```

## 14. Обернена матриця

```python
np.linalg.inv(m)
```

## 15. Розв'язання системи рівнянь

```python
A = np.array([[2,1,-3],[1,3,2],[3,-1,1]])
b = np.array([5,8,2])
np.linalg.solve(A,b)
```

## 16. Множення матриць

```python
M1 = np.matrix([[2,0,-1],[1,3,4]])
M2 = np.matrix([[1,2],[3,1],[-2,0]])
M1 @ M2
```

## 17. Транспонування матриці

```python
M1.T
```

## 18. Симетричність матриці

```python
sym = np.array([[4,1,1],[1,4,1],[1,1,4]])
np.allclose(sym, sym.T)
```

## 19. Слід матриці

```python
np.trace(sym)
```

## 20. Операції над матрицями

```python
K = np.array([[1,2],[3,5]])
L = np.array([[2,1],[4,0]])
K+L, K-L, K*L
```

## 21. Матриця у степені

```python
np.linalg.matrix_power(K,3)
```

## 22. Сортування матриці

```python
T = np.array([[9,2,5],[1,3,7],[6,4,8]])
np.sort(T, axis=0), np.sort(T, axis=1)
```

## 23. Суми рядків

```python
T = np.array([[2,4],[1,6]])
T.sum(axis=1).reshape(-1,1)
```

Закріплено навички виконання лінійних, статистичних та алгебраїчних операцій, а також роботу з випадковими даними та системами рівнянь.
